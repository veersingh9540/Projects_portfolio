## Breast Cancer Classification. 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import sklearn.datasets
```

#### Data Importing and Preprocessing

```python
columns_name = ['ID','DIAGNOSIS','RADIUS_MEAN','TEXTURE_MEAN','PERIMETER_MEAN','AREA_MEAN','SMOOTHNESS_MEAN',
            'COMPACTNESS_MEAN','CONCATIVITY_MEAN','CONCAVE_POINTS_MEAN','SYMMETRY_MEAN','FRACTAL_DIMENSION_MEAN','RADIUS_SE','TEXTURE_SE',
            'PERIMETER_SE', 'AREA_SE', 'SMOOTHNESS_SE', 'COMPACTNESS_SE', 'CONCATIVITY_SE', 'CONCAVE_POINTS_SE', 'SYMMETRY_SE', 'FRACTAL_DIMENSION_SE',
            'RADIUS_WORST', 'TEXTURE_WORST', 'PERIMETER_WORST', 'AREA_WORST', 'SMOOTHNESS_WORST', 'COMPACTNESS_WORST', 'CONCATIVITY_WORST','CONCAVE_POINTS_WORST', 'SYMMETRY_WORST', 'FRACTAL_DIMENSION_WORST'
           ]
```

#### Converting Diagnosis column to 0,1 for training the model.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb_Diagnosis = LabelEncoder()
```

```python
Breast_Data['Label_Diagnosis'] = lb_Diagnosis.fit_transform(Breast_Data['DIAGNOSIS'])
```

#### Created Label_Diagnosis column where M: 1 and B:0

```python
Breast_Data['Label_Diagnosis'].info
```

### Training the Dataset and Splitting into Test and Training variables.

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
X = Breast_Data.drop(columns=['DIAGNOSIS', 'Label_Diagnosis'], axis= 1)
Y = Breast_Data['Label_Diagnosis']
```

```python
X.shape, Y.shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
Y_train
```

#### Scaling the Training and Tsting Dataframe "As we know that remaining columns is a Numerical Data so it would be beneficial for sklearn's regresstion model understand the data while training"

```python
Scaler = preprocessing.StandardScaler()
```

```python
X_trainS = Scaler.fit_transform(X_train)
X_testS = Scaler.fit_transform(X_test)
```

```python
X_trainS.shape, X_testS.shape
```

####  Importing Tensorflow and keras to setup Neural network for classification

```python
import tensorflow as tf 
tf.random.set_seed(3)
```

#### Making neural network layers using keras, Hidden Layer

```python
from tensorflow import keras
MODEL = keras.Sequential([
                          keras.layers.Flatten(input_shape=(31,)),
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(2, activation='sigmoid')
])
```

```python
MODEL.compile(
            optimizer='adam', 
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy']
)
```

#### Training the Neural network 


```python
trained_model = MODEL.fit(X_trainS, Y_train, epochs=25, validation_split=0.01)
```

```python
trained_model.params
```

```python
%matplotlib inline
plt.plot(trained_model.history['accuracy'])
plt.plot(trained_model.history['val_accuracy'])
plt.xlabel('Accuracy vs Epoch')
plt.legend(['Training data', 'Validation data'], loc = 'lower right')
```

```python
%matplotlib inline
plt.plot(trained_model.history['loss'])
plt.plot(trained_model.history['val_loss'])
plt.xlabel('Loss vs Epoch')
plt.legend(['Training data', 'Validation data'], loc = 'upper right')
```

```python
trained_model.history['accuracy']
```

#### Finding The accuracy Of the model


```python
Loss, Accuracy = MODEL.evaluate(X_testS, Y_test)
```

```python
print(Loss, Accuracy)
```

```python
MODEL.predict(X_testS)
```
